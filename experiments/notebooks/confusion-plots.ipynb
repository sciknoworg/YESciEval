```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Data from the tables
orkgsyn_data = np.array([
    [4.85, 4.71, 4.69, 4.80],
    [4.92, 4.89, 4.87, 4.88],
    [4.81, 4.77, 4.75, 4.78],
    [4.84, 4.75, 4.72, 4.81]
])

bioasq_data = np.array([
    [4.82, 3.39, 4.82, 4.82],
    [4.85, 4.48, 4.80, 4.84],
    [4.83, 4.72, 4.80, 4.78],
    [4.82, 3.54, 4.65, 4.78]
])

# Labels for the heatmaps
models = ["Qwen2.5-72B", "LLaMA-3.1-72B", "LLaMA-3.1-8B", "Mistral-Large"]

# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(4, 6))

# ORKGSyn Confusion Matrix
sns.heatmap(orkgsyn_data, annot=True, fmt=".2f", cmap="Blues", xticklabels=models, yticklabels=models, ax=axes[0], annot_kws={"size": 7}, cbar=False)
axes[0].set_title("ORKG-Synthesis", fontsize=9)
# axes[0].set_xlabel("Synthesizer", fontsize=8)
# axes[0].set_ylabel("Evaluator", fontsize=8)
axes[0].tick_params(axis='both', which='major', labelsize=6)
axes[0].tick_params(axis='x', rotation=0)

# BioASQ Confusion Matrix
sns.heatmap(bioasq_data, annot=True, fmt=".2f", cmap="Blues", xticklabels=models, yticklabels=models, ax=axes[1], annot_kws={"size": 7}, cbar=False)
axes[1].set_title("BioASQ", fontsize=9)
# axes[1].set_xlabel("Synthesizer", fontsize=8)
# axes[1].set_ylabel("Evaluator", fontsize=8)
axes[1].tick_params(axis='both', which='major', labelsize=6)
axes[1].tick_params(axis='x', rotation=0)

# Adjust layout and show the plots
plt.tight_layout()
plt.savefig("images/confusion_matrix_vanilla_v2.pdf", format="pdf", bbox_inches="tight")
plt.savefig("images/confusion_matrix_vanilla_v2.png", format="png", dpi=300, bbox_inches="tight")
plt.show()
```
